```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline
```

```python
x = np.arange(-2, 2.5, 0.25)
y = np.arange(-2, 2.5, 0.25)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)
plt.show()
plt.contour(X,Y,Z)
plt.axis('square')
plt.show()
```

```python
Z = (1/5)*X**2 + Y**2
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z)
plt.show()
plt.contour(X,Y,Z)
plt.axis('square')
plt.show()
```

```python
Z = np.sqrt(9-X**2-Y**2)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z)
plt.show()
plt.contour(X,Y,Z)
plt.axis('square')
plt.show()
```

```python
Z = np.sin(X*Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z)
plt.show()
plt.contour(X,Y,Z)
plt.axis('square')
plt.show()
```
